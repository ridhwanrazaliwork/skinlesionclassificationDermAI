### Dataset Overview

The [dataset](https://www.kaggle.com/datasets/cdeotte/jpeg-isic2019-128x128), sourced from Kaggle's SIIM-ISIC Melanoma Classification competition consists of approximately 33,125 JPEG dermatoscopic images of skin lesions with a resolution of 128x128 pixels. Each image is named with a unique identifier (e.g., `ISIC_0027419.jpg`) and are categorized into **9 distinct classes**, representing a range of benign and malignant skin conditions.

### Column Names and Descriptions

| Column Name               | Description                                                                                       |
|--------------------------------|-------------------------------------------------------------------------------------------------------|
| image_name                  | The unique identifier for each image.                     |
| patient_id                  | Anonymized patient identifier.                 |
| sex                         | The gender of the patient (`male`, `female`, or `unknown`).                                          |
| age_approx                  | The approximate age of the patient in years.                                                        |
| anatom_site_general_challenge | The anatomical site of the lesion (torso, head/neck, upper extremity, lower extremity, oral/genital, palms/soles, unknown).  |
| diagnosis                   | The specific diagnosis of the lesion.          |
| benign_malignant            | Indicates whether the lesion is benign or malignant. Redundant with target.                       |
| target                      | The binary classification target: 1 for malignant (melanoma), 0 for benign.                     |
| tfrecords                   | The TFRecord shard identifier where the image is stored. Used for TensorFlow pipelines.             |
| height                      | The original height of the image before resizing.                                                   |
| width                       | The original width of the image before resizing.                                                    |


### Classes and Their Descriptions


| Class Label                         | Description                                                                 |
|-------------------------------------|-----------------------------------------------------------------------------|
| melanoma                          | A serious type of skin cancer that can spread to other body parts.          |
| nevus                             | Common moles that are usually benign.                                       |
| atypical melanocytic proliferation| Atypical lesions that may indicate a risk of melanoma.                      |
| cafe-au-lait macule               | Benign light brown patches on the skin.                                     |
| lentigo NOS                       | Non-specific lentigo, often benign pigmented spots on the skin.             |
| lichenoid keratosis               | Benign skin lesions with inflammatory and keratotic changes.                |
| seborrheic keratosis              | Non-cancerous skin growth, often appearing as brown or black bumps.         |
| solar lentigo                     | Benign pigmented spots caused by sun exposure, common in older adults.      |
| unknown                           | Unspecified diagnosis, indicating ambiguity in classification.              |

In order to gain a deeper understanding on the dataset structure, the columns are categorized into three primary features based on the type of data they represent, which are:

| Key Features                         | Types of Data                                                                 |
|-------------------------------------|-----------------------------------------------------------------------------|
| Demographic                          | age_approx, sex          |
| Skin Lesions                             | anatom_site_general_challenge, diagnosis                                       |
| Target| 0: Benign, 1: Malignant                      |

### 5.2.3 Define Custom Model CNN (Image only)

```python
# Define Basic CNN
class BasicCNN(nn.Module):
    def __init__(self):
        super(BasicCNN, self).__init__()
        self.conv1 = nn.Conv2d(3, 16, kernel_size=3, stride=1, padding=1)
        self.bn1 = nn.BatchNorm2d(16)
        self.pool1 = nn.MaxPool2d(kernel_size=2, stride=2)

        self.conv2 = nn.Conv2d(16, 32, kernel_size=3, stride=1, padding=1)
        self.bn2 = nn.BatchNorm2d(32)
        self.pool2 = nn.MaxPool2d(kernel_size=2, stride=2)

        # Flatten size calculation. Output size = (128/2/2), 32 x 32 for 2 pooling layers
        self.flatten_size = 32 * 32 * 32 # corrected flattening size
        self.fc1 = nn.Linear(self.flatten_size, 64)
        self.dropout = nn.Dropout(0.3) # Dropout added
        self.fc2 = nn.Linear(64, 2)

    def forward(self, x):
        x = self.pool1(F.leaky_relu(self.bn1(self.conv1(x)), negative_slope=0.1))
        x = self.pool2(F.leaky_relu(self.bn2(self.conv2(x)), negative_slope=0.1))
        x = x.view(x.size(0), -1)
        x = self.dropout(F.leaky_relu(self.fc1(x), negative_slope=0.1))
        x = self.fc2(x)
        return x
```

### 5.2.4 Define Model EfficientNet (Image only)

```python
class GeffnetModel(nn.Module): 
    def __init__(self, model_name='efficientnet_b0', num_classes=2, pretrained=True): # Added other options her
        super(GeffnetModel, self).__init__()
        self.model = geffnet.create_model(model_name, pretrained=pretrained)
        num_ftrs = self.model.fc.in_features if hasattr(self.model, 'fc') else self.model.classifier.in_features
        if hasattr(self.model, 'fc'):
            self.model.fc = nn.Identity()  # Remove the original classifier
        else:
            self.model.classifier = nn.Identity()
        self.dropout = nn.Dropout(0.)
        self.fc = nn.Linear(num_ftrs, num_classes)
    
    def forward(self, x):
        x = self.model(x)
        x = self.dropout(x)
        x = self.fc(x)
        return x
```

### 5.2.5 Training functions for Model CNN (Image only)

**Prepares the model for training.**

1. Loads data in batches from the training set.
2. Performs forward pass to get the model output, and calculates the loss based on the prediction.
3. Backpropagates through the network to update the model parameters.
4. Calculates and returns the average loss and the accuracy for the epoch.

```python
def train_epoch(model, data_loader, optimizer, criterion, device):
    model.train()
    total_loss = 0.0
    all_labels = []
    all_predictions = []

    for images, labels in tqdm(data_loader, desc='Training'): # To see our progress here
        images, labels = images.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        _, predicted_labels = torch.max(outputs, 1)

        all_labels.extend(labels.cpu().numpy())
        all_predictions.extend(predicted_labels.cpu().numpy())

    avg_loss = total_loss / len(data_loader)
    acc = accuracy_score(all_labels, all_predictions)

    return avg_loss, acc
```

### 5.2.6 Validation functions for Model CNN (Image only)

1. Performs a forward pass through the model, and calculates the loss.
2. Collects the predictions and true labels, and calculates various evaluation metrics like accuracy, AUC ROC, precision, recall, F1 score and also the confusion matrix.
3. It then prints these metrics after final testing, shows a confusion matrix plot, and returns all of these metrics. (For every epochs)

```python
def validate_epoch(model, loader, criterion, device, model_name, verbose=False):
    model.eval()
    loss_total = 0.0
    correct = 0
    total = 0
    all_labels = []
    all_preds = []

    with torch.no_grad():
        for inputs, labels in loader:
            inputs, labels = inputs.to(device), labels.to(device)

            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss_total += loss.item()

            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

            all_labels.extend(labels.cpu().numpy())
            all_preds.extend(predicted.cpu().numpy())

    accuracy = correct / total
    auc_roc = roc_auc_score(all_labels, all_preds)
    precision = precision_score(all_labels, all_preds)
    recall = recall_score(all_labels, all_preds)
    f1 = f1_score(all_labels, all_preds)
    conf_matrix = confusion_matrix(all_labels, all_preds)
    class_report = classification_report(all_labels, all_preds, target_names=["Benign", "Malignant"], zero_division=0)

    if verbose:
        # Print metrics
        print(f"\nDetailed Evaluation for {model_name}:")
        print(f"  Loss: {loss_total / len(loader):.4f}")
        print(f"  Accuracy: {accuracy:.4f}")
        print(f"  AUC ROC: {auc_roc:.4f}")
        print(f"  Precision: {precision:.4f}")
        print(f"  Recall: {recall:.4f}")
        print(f"  F1-Score: {f1:.4f}")
        print("  Confusion Matrix:")
        print(f"    True Negative (TN): {conf_matrix[0, 0]}")
        print(f"    False Positive (FP): {conf_matrix[0, 1]}")
        print(f"    False Negative (FN): {conf_matrix[1, 0]}")
        print(f"    True Positive (TP): {conf_matrix[1, 1]}")

        # Plot confusion matrix
        plt.figure(figsize=(6, 5))
        sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues",
                    xticklabels=["Non-Cancer", "Cancer"],
                    yticklabels=["Non-Cancer", "Cancer"])
        plt.ylabel("True Label")
        plt.xlabel("Predicted Label")
        plt.title(f'{model_name} Confusion Matrix')
        plt.show()

        # Print classification report
        print(f"\n{model_name} Classification Report:\n{class_report}")

    return (loss_total / len(loader), accuracy, auc_roc, precision, recall, f1, conf_matrix)
```

### 5.2.7 Define function to save models and load models

```python
def save_traced_model(model, model_name, device):
    model.eval()
    example_input = torch.randn(1, 3, 128, 128).to(device) # Modified for 128x128, can be made dynamic if needed.
    traced_model = torch.jit.trace(model, example_input)
    traced_model.save(f"best_{model_name}_model_traced.pt")
    print(f"{model_name} Model and TorchScript Model saved")

def load_traced_model(model_path, device):
    return torch.jit.load(model_path, map_location=device)
```

### 5.2.9 Start the training for Model CNN (Image only)

1. Sets up a training loop that iterates for a defined number of epochs

2. Calls train_epoch to train for an epoch, and then calls validate_epoch to evaluate the model after training for that epoch.

3. Logs the training and validation performance metrics.

4. Dynamically updates the learning rate using the ReduceLROnPlateau scheduler.

5. Saves the best model based on the AUC ROC, and also sets up an early stopping approach to reduce overfitting and save training time.

```python

for model_name, model_class in cnn_architectures.items():
    print("=" * 40)
    print(f"Training: {model_name}")
    print("=" * 40)

    # Setup
    cnn_model = model_class().to(device)
    criterion_cnn = nn.CrossEntropyLoss(weight=class_weights).to(device)
    optimizer_cnn = optim.Adam(cnn_model.parameters(), lr=learning_rate_init)
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer_cnn, mode='max', factor=0.5, patience=patience)

    best_val_loss_cnn = float('inf')
    best_val_auc_roc_cnn = 0.0
    best_cnn_metrics = None
    history_cnn = {'loss': [], 'val_loss': [], 'accuracy': [], 'val_accuracy': [], 'val_auc_roc': []}
    trigger = 0

    # Training Loop
    for epoch in range(epochs):
        train_loss, train_acc = train_epoch(cnn_model, train_loader, optimizer_cnn, criterion_cnn, device)

        val_loss, val_acc, val_auc_roc, val_precision, val_recall, val_f1, val_conf_matrix = validate_epoch(
            cnn_model, val_loader, criterion_cnn, device, model_name, verbose=False
        )
        print(f"Epoch {epoch+1}/{epochs}: "
              f"Train Loss: {train_loss:.4f}, Train Accuracy: {train_acc:.4f}, "
            f"Val Loss: {val_loss:.4f}, Val Accuracy: {val_acc:.4f}, Val AUC ROC: {val_auc_roc:.4f}")

        history_cnn['loss'].append(train_loss)
        history_cnn['val_loss'].append(val_loss)
        history_cnn['accuracy'].append(train_acc)
        history_cnn['val_accuracy'].append(val_acc)
        history_cnn['val_auc_roc'].append(val_auc_roc)

        scheduler.step(val_auc_roc)
        current_lr = scheduler.get_last_lr()[0]
        print(f"  Current Learning Rate: {current_lr:.6f}")


        if val_auc_roc > best_val_auc_roc_cnn:
            best_val_auc_roc_cnn = val_auc_roc
            best_cnn_metrics = {"val_loss":val_loss, "val_acc":val_acc, "val_auc_roc":val_auc_roc,
                                "precision":val_precision, "recall":val_recall, "f1":val_f1,
                                "conf_matrix":val_conf_matrix}
            # Saved the model for usecase later
            # We save every 'best' model based on auc_roc
            torch.save(cnn_model.state_dict(), f"best_{model_name}_model_weight_only.pth")
            # We save every 'best' model weight + architecture based on auc_roc
            torch.save(cnn_model, f"best_{model_name}_model_full.pth")
            
            save_traced_model(cnn_model, model_name, device)
            print(f"{model_name} Model and TorchScript Model saved")
            trigger=0
        else:
            trigger+=1
        if trigger>= patience:
            print(f"Early stopping triggered after {epoch +1} epochs")
            break

    # After the training loop ends, we now plot the evaluation results (Test evaluation)
    print(f'{model_name} Final Evaluation:')  # We print after test evaluation

    # Perform final evaluation with test set
    # print(f'\nTest Evaluation for {model_name} based on Best Validation AUC ROC Model:')
    cnn_model.load_state_dict(torch.load(f"best_{model_name}_model_weight_only.pth"))  # Load best model weight
    cnn_model.eval()  # Set the model to evaluation mode

    print(f'\nTest Evaluation for {model_name} based on Best Validation AUC ROC Model:')
    test_loss, test_acc, test_auc_roc, test_precision, test_recall, test_f1, test_conf_matrix = validate_epoch(
    cnn_model, test_loader, criterion_cnn, device, model_name, verbose=True
    )
    print(f'  Test Loss: {test_loss:.4f}')
    print(f'  Test Accuracy: {test_acc:.4f}')
    print(f'  Test AUC ROC: {test_auc_roc:.4f}')

    # Plotting results after test evaluation
    plt.style.use('seaborn-v0_8-pastel')

    # Plot CNN model results
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    # Plot Loss Curves
    epochs_range = range(1, len(history_cnn['loss']) + 1)
    ax1.plot(epochs_range, history_cnn['loss'], 'b-', label='Training Loss', linewidth=2)
    ax1.plot(epochs_range, history_cnn['val_loss'], 'r-', label='Validation Loss', linewidth=2)
    ax1.set_title(f'{model_name} Loss Curves', fontsize=12, pad=15)
    ax1.set_xlabel('Epoch', fontsize=10)
    ax1.set_ylabel('Loss', fontsize=10)
    ax1.legend(fontsize=10)
    ax1.grid(True)

    # Plot Accuracy Curves
    ax2.plot(epochs_range, history_cnn['accuracy'], 'b-', label='Training Accuracy', linewidth=2)
    ax2.plot(epochs_range, history_cnn['val_accuracy'], 'r-', label='Validation Accuracy', linewidth=2)
    ax2.set_title(f'{model_name} Accuracy Curves', fontsize=12, pad=15)
    ax2.set_xlabel('Epoch', fontsize=10)
    ax2.set_ylabel('Accuracy', fontsize=10)
    ax2.legend(fontsize=10)
    ax2.grid(True)

    plt.tight_layout()
    plt.show()

    # Plot AUC ROC Curve
    plt.figure(figsize=(8, 6))
    plt.plot(epochs_range, history_cnn['val_auc_roc'], 'g-', label='Validation AUC ROC', linewidth=2)
    plt.title(f'{model_name} Validation AUC ROC Curve', fontsize=12, pad=15)
    plt.xlabel('Epoch', fontsize=10)
    plt.ylabel('AUC ROC', fontsize=10)
    plt.legend(fontsize=10)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

4. Prepares the images, the metadata, and labels as PyTorch tensors for training and testing the CNN + Metadata model.

###  5.3.1 Create MelanomaMetadataDataset class for Model CNN (Image and metadata)

```python
# # MelanomaMetadataDataset for Image with Metadata
class MelanomaMetadataDataset(Dataset):
    def __init__(self, dataframe, train_transform=None, val_transform=None):
        self.dataframe = dataframe
        self.train_transform = train_transform
        self.val_transform = val_transform
        self.metadata_cols = ['age_encoded', 'sex_encoded', 'anatom_encoded']

    def __len__(self):
        return len(self.dataframe)

    def __getitem__(self, idx):
        img_path = self.dataframe.iloc[idx]['image_path']
        image = Image.open(img_path).convert('RGB')
        label = self.dataframe.iloc[idx]['target']

        if self.train_transform:
            image = self.train_transform(image)
        else:
            image = self.val_transform(image)

        metadata_values = []
        for col in self.metadata_cols:
           if col == 'age_encoded':
              metadata_values.append(self.dataframe.iloc[idx][col])
           else:
              metadata_values.extend(self.dataframe.iloc[idx][col])

        metadata = np.array(metadata_values,dtype=np.float32) # Convert metadata into numpy first
        metadata = torch.tensor(metadata, dtype=torch.float32) # Then convert metadata into tensor

        return image, metadata, torch.tensor(label, dtype=torch.long)



train_dataset_metadata = MelanomaMetadataDataset(train_data, train_transform=train_transform)
val_dataset_metadata = MelanomaMetadataDataset(val_data, val_transform = val_transform)
test_dataset_metadata = MelanomaMetadataDataset(test_data, val_transform = val_transform)
```

### 5.3.2 The DataLoader class in PyTorch is used for streamlining the process of loading data.

```python
train_loader_metadata = DataLoader(train_dataset_metadata, batch_size=batch_size, shuffle=True, num_workers=0)
val_loader_metadata = DataLoader(val_dataset_metadata, batch_size=batch_size, shuffle=False, num_workers=0)
test_loader_metadata = DataLoader(test_dataset_metadata, batch_size=batch_size, shuffle=False, num_workers=0)
```

### 5.3.3 Define Model CNN (Image and metadata)

```python
# Define CNN with Metadata Model
class CNNImageMetadata(nn.Module):
    def __init__(self, num_metadata_features, num_sex_categories, num_anatom_categories):
        super(CNNImageMetadata, self).__init__()
        # Image processing branch
        self.conv1 = nn.Conv2d(3, 16, kernel_size=3, stride=1, padding=1)
        self.relu1 = nn.ReLU()
        self.pool1 = nn.MaxPool2d(kernel_size=2, stride=2)
        self.conv2 = nn.Conv2d(16, 32, kernel_size=3, stride=1, padding=1)
        self.relu2 = nn.ReLU()
        self.pool2 = nn.MaxPool2d(kernel_size=2, stride=2)

        # Calculate the flattened size
        self.flatten_size = 32 * 32 * 32

        # Image feature reduction
        self.fc_image = nn.Sequential(
            nn.Linear(self.flatten_size, 64),
            nn.ReLU()
        )

        # Metadata processing branch
        self.fc_metadata = nn.Sequential(
            nn.Linear(num_metadata_features, 32),
            nn.ReLU(),
            nn.Linear(32, 32),
            nn.ReLU()
        )

        # Combining branches
        self.fc_combine = nn.Sequential(
            nn.Linear(64 + 32, 64),
            nn.ReLU(),
            nn.Linear(64, 2)
        )

    def forward(self, image, metadata):
        # Process image
        x = self.pool1(self.relu1(self.conv1(image)))
        x = self.pool2(self.relu2(self.conv2(x)))
        x = x.view(-1, self.flatten_size)  # Flatten
        image_features = self.fc_image(x)

        # Process metadata
        metadata_features = self.fc_metadata(metadata)

        # Combine features
        combined = torch.cat((image_features, metadata_features), dim=1)
        output = self.fc_combine(combined)

        return output
```

### 5.3.4 Training functions for Model CNN (Image and metadata)

**Prepares the model for training.**

1. Loads data in batches from the training set.
2. Performs forward pass to get the model output, and calculates the loss based on the prediction.
3. Backpropagates through the network to update the model parameters.
4. Calculates and returns the average loss and the accuracy for the epoch.

```python
# Training functions (metadata)
def train_epoch_metadata(model, data_loader, optimizer, criterion, device):
    model.train()
    total_loss = 0.0
    all_labels = []
    all_predictions = []

    for images, metadata, labels in tqdm(data_loader, desc='Training'):
        # print(images.shape, metadata.shape, labels.shape)
        images, metadata, labels = images.to(device), metadata.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(images, metadata)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        _, predicted_labels = torch.max(outputs, 1)

        all_labels.extend(labels.cpu().numpy())
        all_predictions.extend(predicted_labels.cpu().numpy())

    avg_loss = total_loss / len(data_loader)
    acc = accuracy_score(all_labels, all_predictions)

    return avg_loss, acc
```

### 5.3.5 Validation functions for Model CNN (Image and metadata)


1. Performs a forward pass through the model, and calculates the loss.
2. Collects the predictions and true labels, and calculates various evaluation metrics like accuracy, AUC ROC, precision, recall, F1 score and also the confusion matrix.
3. It then prints these metrics, shows a confusion matrix plot, and returns all of these metrics. (For every epochs)

```python
def validate_epoch_metadata(model, data_loader, criterion, device, model_name, verbose=False):
    model.eval()
    total_loss = 0.0
    all_labels = []
    all_predictions = []
    all_probs = []

    with torch.no_grad():
        for images, metadata, labels in tqdm(data_loader, desc="Validation"):
            images, metadata, labels = images.to(device), metadata.to(device), labels.to(device)
            outputs = model(images, metadata)
            loss = criterion(outputs, labels)
            total_loss += loss.item()
            _, predicted_labels = torch.max(outputs, 1)
            all_labels.extend(labels.cpu().numpy())
            all_predictions.extend(predicted_labels.cpu().numpy())
            probs = torch.softmax(outputs, dim=1)[:, 1].cpu().numpy()
            all_probs.extend(probs)


    avg_loss = total_loss / len(data_loader)
    acc = accuracy_score(all_labels, all_predictions)
    auc_roc = roc_auc_score(all_labels, all_probs)
    conf_matrix = confusion_matrix(all_labels, all_predictions)
    precision = precision_score(all_labels, all_predictions, zero_division = 0)
    recall = recall_score(all_labels, all_predictions, zero_division = 0)
    f1 = f1_score(all_labels, all_predictions, zero_division = 0)
    class_report = classification_report(all_labels, all_predictions, target_names=["Benign", "Malignant"], zero_division=0)

    if verbose:
        # Print metrics nicely
        print(f"\n{model_name} Validation Metrics:")
        print(f"  Validation Loss: {avg_loss:.4f}")
        print(f"  Validation Accuracy: {acc:.4f}")
        print(f"  Validation AUC ROC: {auc_roc:.4f}")
        print(f"  Precision (True Positives): {precision:.4f}") # How many correct predictions of the positive class
        print(f"  Recall (Sensitivity): {recall:.4f}") # How well it predicted the positive class
        print(f"  F1-Score: {f1:.4f}")  # Harmonic mean of precision and recall

        print("  Confusion Matrix:")
        print(f"    True Negative (TN): {conf_matrix[0, 0]}")
        print(f"    False Positive (FP): {conf_matrix[0, 1]}")
        print(f"    False Negative (FN): {conf_matrix[1, 0]}")
        print(f"    True Positive (TP): {conf_matrix[1, 1]}")

        # Plot the confusion matrix
        plt.figure(figsize=(6, 5))
        sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues",
                    xticklabels=["Non-Cancer", "Cancer"],
                    yticklabels=["Non-Cancer", "Cancer"])
        plt.ylabel("True Label")
        plt.xlabel("Predicted Label")
        plt.title(f'{model_name} Confusion Matrix')
        plt.show()

        # Print classification report
        print(f"\n{model_name} Classification Report:\n{class_report}")


    return avg_loss, acc, auc_roc, precision, recall, f1, conf_matrix
```

### 5.3.6 Setup for Model CNN (Image and metadata)

```python
dnum_metadata_features = len(metadata_cols)  # Should be number of metadata features (age + one-hot encoded categories)
num_sex_categories = len(sex_categories)
num_anatom_categories = len(anatom_categories)

# Setup
cnn_metadata_model = CNNImageMetadata(
    num_metadata_features=num_sex_categories + num_anatom_categories + 1,
    num_sex_categories=num_sex_categories,
    num_anatom_categories=num_anatom_categories
).to(device)

criterion_cnn_metadata = nn.CrossEntropyLoss(weight=class_weights).to(device)
optimizer_cnn_metadata = optim.Adam(cnn_metadata_model.parameters(), lr=learning_rate_init)
scheduler_cnn_metadata = optim.lr_scheduler.ReduceLROnPlateau(optimizer_cnn_metadata, mode='max', factor=0.5, patience=patience)

best_val_loss_cnn_metadata = float('inf')
best_val_auc_roc_cnn_metadata = 0.0
best_cnn_metadata_metrics = None
history_cnn_metadata = {'loss': [], 'val_loss': [], 'accuracy': [], 'val_accuracy': [], 'val_auc_roc': []}
trigger = 0
```

### Start the training for Model CNN (Image and metadata)

```python
# Training Loop
for epoch in range(epochs):
    train_loss, train_acc = train_epoch_metadata(
        cnn_metadata_model,
        train_loader_metadata,
        optimizer_cnn_metadata,
        criterion_cnn_metadata,
        device
    )

    val_loss, val_acc, val_auc_roc, val_precision, val_recall, val_f1, val_conf_matrix = validate_epoch_metadata(
        cnn_metadata_model,
        val_loader_metadata,
        criterion_cnn_metadata,
        device,
        'cnn_metadata', 
        verbose=False
    )
    
    # Store metrics
    history_cnn_metadata['loss'].append(float(train_loss))
    history_cnn_metadata['val_loss'].append(float(val_loss))
    history_cnn_metadata['accuracy'].append(float(train_acc))
    history_cnn_metadata['val_accuracy'].append(float(val_acc))
    history_cnn_metadata['val_auc_roc'].append(float(val_auc_roc))

    print(f"Epoch {epoch+1}/{epochs}: "
          f"Train Loss: {train_loss:.4f}, Train Accuracy: {train_acc:.4f}, "
          f"Val Loss: {val_loss:.4f}, Val Accuracy: {val_acc:.4f}, Val AUC ROC: {val_auc_roc:.4f}")

    # Learning Rate Scheduler Step here
    scheduler_cnn_metadata.step(val_auc_roc) # changed from `val_loss` to `val_auc_roc`
    current_lr = scheduler_cnn_metadata.get_last_lr()[0] # Get the learning rate here
    print(f"  Current Learning Rate: {current_lr:.6f}")  # log learning rate


    # Early Stopping Implementation
    if val_auc_roc > best_val_auc_roc_cnn_metadata:
        best_val_auc_roc_cnn_metadata = val_auc_roc
        best_cnn_metadata_metrics = {
            "val_loss": val_loss,
            "val_acc": val_acc,
            "val_auc_roc": val_auc_roc,
            "precision": val_precision,
            "recall": val_recall,
            "f1": val_f1,
            "conf_matrix": val_conf_matrix
        }
        # We save every 'best' model weight based on auc_roc
        torch.save(cnn_metadata_model.state_dict(), f"best_CNNImageMetadata_cnn_metadata_model_weight_only.pth")
        # We save every 'best' model weight + architecture based on auc_roc
        torch.save(cnn_metadata_model, f"best_CNNImageMetadata_cnn_metadata_model_full.pth")
        # Save traced model using TorchScript
        example_image_input = torch.randn(1, 3, 128, 128).to(device)  # Example input size for now
        example_metadata_input = torch.randn(1, num_sex_categories + num_anatom_categories + 1).to(device)  # Example metadata input
        # Perform JIT tracing
        cnn_metadata_model.eval()  # Set the model to evaluation mode
        traced_model = torch.jit.trace(cnn_metadata_model, (example_image_input, example_metadata_input))
        traced_model.save(f"best_CNNImageMetadata_cnn_metadata_model_traced.pt")

        print(f"CNNImageMetadata CNN_Metadata Model and TorchScript Traced Model saved")
        trigger=0

    else:
        trigger+=1

    if trigger>= patience:
        print(f"Early stopping triggered after {epoch +1} epochs")
        break

# After the training loop ends, we now plot the evaluation results (Test evaluation)
print(f'CNNImageMetadata Final Evaluation:')  # We print after test evaluation

# Perform final evaluation with test set
# print(f'\nTest Evaluation for CNNImageMetadata based on Best Validation AUC ROC Model:')
cnn_metadata_model.load_state_dict(torch.load(f"best_CNNImageMetadata_cnn_metadata_model_weight_only.pth"))  # Load best model weight
cnn_metadata_model.eval()  # Set the model to evaluation mode

test_loss, test_acc, test_auc_roc, test_precision, test_recall, test_f1, test_conf_matrix = validate_epoch_metadata(
        cnn_metadata_model,
        test_loader_metadata,
        criterion_cnn_metadata,
        device,
        'cnn_metadata',
        verbose=True # Set this to true for final eval
    )
print(f'  Test Loss: {test_loss:.4f}')
print(f'  Test Accuracy: {test_acc:.4f}')
print(f'  Test AUC ROC: {test_auc_roc:.4f}')

# Plotting with seaborn style
plt.style.use('seaborn-v0_8-pastel')

# Plot CNN metadata model results
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
# Plot Loss Curves
epochs_range = range(1, len(history_cnn_metadata['loss']) + 1)
ax1.plot(epochs_range, history_cnn_metadata['loss'], 'b-', label='Training Loss', linewidth=2)
ax1.plot(epochs_range, history_cnn_metadata['val_loss'], 'r-', label='Validation Loss', linewidth=2)
ax1.set_title(f'CNNImageMetadata Loss Curves', fontsize=12, pad=15)
ax1.set_xlabel('Epoch', fontsize=10)
ax1.set_ylabel('Loss', fontsize=10)
ax1.legend(fontsize=10)
ax1.grid(True)

# Plot Accuracy Curves
ax2.plot(epochs_range, history_cnn_metadata['accuracy'], 'b-', label='Training Accuracy', linewidth=2)
ax2.plot(epochs_range, history_cnn_metadata['val_accuracy'], 'r-', label='Validation Accuracy', linewidth=2)
ax2.set_title(f'CNNImageMetadata Accuracy Curves', fontsize=12, pad=15)
ax2.set_xlabel('Epoch', fontsize=10)
ax2.set_ylabel('Accuracy', fontsize=10)
ax2.legend(fontsize=10)
ax2.grid(True)

plt.tight_layout()
plt.show()

# Plot AUC ROC Curve
plt.figure(figsize=(8, 6))
plt.plot(epochs_range, history_cnn_metadata['val_auc_roc'], 'g-', label='Validation AUC ROC', linewidth=2)
plt.title(f'CNNImageMetadata Validation AUC ROC Curve', fontsize=12, pad=15)
plt.xlabel('Epoch', fontsize=10)
plt.ylabel('AUC ROC', fontsize=10)
plt.legend(fontsize=10)
plt.grid(True)
plt.tight_layout()
plt.show()
```

### 6.2 Grad CAM for Model CNN (Image only)
Grad-CAM works best for image data because it relies on convolutional layers to generate visual explanations which is why Grad CAM applies over Model CNN (Image only). <br>
When metadata is added, it doesn’t directly affect these layers, but it can complicate the interpretation since Grad-CAM can’t account for non-image features.

```python
import torch
import torch.nn as nn
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
import random
from PIL import Image

def show_cam_on_image(img, mask, use_rgb=False, intensity=0.5):
    heatmap = cv2.applyColorMap(np.uint8(255 * mask), cv2.COLORMAP_JET)
    heatmap = np.float32(heatmap) / 255
    if use_rgb:
        heatmap = cv2.cvtColor(heatmap, cv2.COLOR_BGR2RGB)
    # Multiply heatmap by intensity
    cam = heatmap * intensity + img
    return np.uint8(255 * cam)

batch_size = 1
input_shape = (3, 128, 128)

# Select the device (CPU or GPU)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = BasicCNN().to(device)

# Load a random image from the train folder
train_folder_path = '/kaggle/input/jpeg-isic2019-128x128/train/'
all_files = os.listdir(train_folder_path)
random_file = random.choice(all_files)
img_path = os.path.join(train_folder_path, random_file)

img_pil = Image.open(img_path).convert('RGB')

# Preprocess the image
input_tensor = val_transform(img_pil)
input_tensor = input_tensor.unsqueeze(0).to(device)

# Set model to evaluation mode
model.eval()

# 1. Define a variable to store the activations of the chosen layer
middle_layer_activations = None

# 2. Define a hook function
def hook_fn(module, input, output):
  global middle_layer_activations
  middle_layer_activations = output
  middle_layer_activations.requires_grad_(True)
   # Retain the gradients on the output tensor
  middle_layer_activations.retain_grad()

# 3. Forward pass to get prediction and activations
target_layer = model.pool2
hook = target_layer.register_forward_hook(hook_fn)
output = model(input_tensor)

# 4. Get predicted class (for Grad-CAM)
predicted_class = torch.argmax(output, dim=1).item()

# Calculate gradient
labels = None
if labels is not None:
    target_label = labels
else:
    target_label = predicted_class
output[0,target_label].backward()

# Get the gradients and activations
if middle_layer_activations is not None:
  gradients = middle_layer_activations.grad.data
else:
  raise Exception("middle_layer_activations is none")

# 5. Make sure middle_layer_activations is not None
if middle_layer_activations is None:
  raise Exception("Middle layer activations are not available, ensure the hook is being applied in the correct layer.")

# 6. Remove the hook
hook.remove()
pooled_gradients = torch.mean(gradients, dim=[0, 2, 3])

# 7. Weight activations by gradients to get the heatmap
activations = middle_layer_activations.detach().cpu().numpy()[0]
for i in range(activations.shape[0]):
    activations[i, :, :] *= pooled_gradients[i].cpu().numpy()
heatmap = np.mean(activations, axis=0)

heatmap = np.maximum(heatmap, 0)
if np.max(heatmap) != 0:
   heatmap /= np.max(heatmap)

# 8. Resize heatmap to match original image size 2D
heatmap = cv2.resize(heatmap, (img_pil.size[0], img_pil.size[1]))
print("Shape of Heatmap after resizing", heatmap.shape)

# Overlay the heatmap on the original image
img = np.array(img_pil, dtype=np.float32)
visualization = show_cam_on_image(img, heatmap, use_rgb=True, intensity=0.8)

# 9. Display original image and Grad-CAM visualization side by side
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Original Image
axs[0].imshow(np.array(img_pil)) # Convert the PIL image to numpy for plotting
axs[0].text(0.5, 1.1, random_file, size=12, ha="center", transform=axs[0].transAxes)
axs[0].set_title("Original Image")
axs[0].axis('off')

# Grad-CAM Visualization
axs[1].imshow(visualization)
axs[1].set_title("Grad-CAM Visualization")
axs[1].axis('off')

plt.show()
```
